```python
# 📦 Import xlwings — ควบคุมแอป Excel โดยตรงจาก Python
# ต่างจาก openpyxl (ทำงานกับไฟล์) xlwings สื่อสารกับแอป Excel ที่เปิดอยู่จริง
import xlwings as xw
```

## 1. สร้าง Workbook Object
เปิดหรือสร้าง Excel workbook

```python
# 📗 สร้าง workbook ใหม่ — จะเปิด Excel และสร้างไฟล์เปล่า
# xw.Book() สร้างและเปิด workbook ใน Excel พร้อมกัน
wb = xw.Book()

# 🔍 ตรวจสอบชื่อ workbook (จะเป็นชื่อเช่น 'Book1' จนกว่าจะบันทึก)
print(f"ชื่อ Workbook: {wb.fullname}")
```

```python
# 📋 ตรวจสอบจำนวน worksheet ใน workbook
# workbook ใหม่จะมี 1 sheet ชื่อ 'Sheet1' เป็นค่าเริ่มต้น
print(f"Worksheet ทั้งหมด: {wb.sheets}")
print(f"จำนวน sheet: {len(wb.sheets)}")
```

## 2. เพิ่ม Worksheet

```python
# ➕ เพิ่ม worksheet ใหม่พร้อมตั้งชื่อ
# sheets.add() สร้าง sheet ใหม่ — จะอยู่ตำแหน่งแรกเป็นค่าเริ่มต้น
wb.sheets.add(name='Test')
print(f"Sheet หลังเพิ่ม 'Test': {[s.name for s in wb.sheets]}")
```

```python
# ➕ เพิ่ม worksheet ก่อน sheet ที่ระบุ
# before='Sheet1' หมายถึง sheet ใหม่จะอยู่ทางซ้ายของ Sheet1
wb.sheets.add(name='Test2', before='Sheet1')
print(f"Sheet ทั้งหมด: {[s.name for s in wb.sheets]}")
```

```python
# ➕ เพิ่ม worksheet หลัง sheet ที่ระบุ
# after='Sheet1' หมายถึง sheet ใหม่จะอยู่ทางขวาของ Sheet1
wb.sheets.add(name='Test3', after='Sheet1')
print(f"Sheet ทั้งหมด: {[s.name for s in wb.sheets]}")
```

## 3. เพิ่ม Worksheet โดยใช้ Input จากผู้ใช้

```python
# 🎤 เพิ่ม worksheet ด้วยชื่อที่ผู้ใช้กำหนดโดยใช้ input()
# input() หยุดการทำงานและรอให้ผู้ใช้พิมพ์ชื่อ
sheet_name = input("กรุณาป้อนชื่อ sheet ใหม่: ")
if sheet_name:
    wb.sheets.add(name=sheet_name)
    print(f"เพิ่ม sheet แล้ว: {sheet_name}")
    print(f"Sheet ทั้งหมด: {[s.name for s in wb.sheets]}")
```

## 4. สร้าง Object ของ Worksheet

```python
# 🎯 เข้าถึง worksheet ด้วยชื่อ — ได้ sheet OBJECT ที่สามารถจัดการได้
sht = wb.sheets['Test']
print(f"Sheet ตามชื่อ: {sht.name}")

# 🎯 เข้าถึงด้วย INDEX (เริ่มจาก 0 เหมือน Python list)
# sheets[0] = sheet แรก, sheets[1] = sheet ที่สอง, เป็นต้น
sht_by_index = wb.sheets[0]
print(f"Sheet แรก (index 0): {sht_by_index.name}")
```

## 5. เปลี่ยนชื่อ Worksheet

```python
# ✏️ เปลี่ยนชื่อ worksheet — แค่กำหนดค่า .name property
# การเปลี่ยนจะเกิดขึ้นทันทีใน Excel
sht.name = 'RenamedSheet'
print(f"Sheet ที่เปลี่ยนชื่อ: {sht.name}")
print(f"Sheet ทั้งหมด: {[s.name for s in wb.sheets]}")
```

## 6. ซ่อน / แสดง Worksheet

```python
# 👁️ ซ่อน worksheet — sheet ยังอยู่แต่ไม่แสดงใน Excel
# ใช้ .visible property ของ xlwings (ใช้ได้ทั้ง macOS และ Windows)
# หมายเหตุ: ซ่อนทุก sheet ไม่ได้ — ต้องมีอย่างน้อย 1 sheet ที่มองเห็น
wb.sheets['Sheet1'].visible = False
print("ซ่อน 'Sheet1' แล้ว")
```

```python
# 👁️ แสดง worksheet — ทำให้ sheet ที่ซ่อนมองเห็นได้อีกครั้ง
# แค่กำหนด .visible กลับเป็น True
wb.sheets['Sheet1'].visible = True
print("แสดง 'Sheet1' อีกครั้งแล้ว")
```

## 7. คัดลอก / ย้าย Worksheet

```python
# 📋 คัดลอก worksheet — สร้างสำเนาของ sheet ที่มีอยู่
# .copy() กับ before= จะวางสำเนาไว้ก่อน sheet ที่ระบุ
# sheet ที่คัดลอกจะได้ชื่ออัตโนมัติเช่น "RenamedSheet (2)"
# มีประโยชน์สำหรับสร้างสำรองหรือเทมเพลตจาก sheet ที่มีอยู่
wb.sheets['RenamedSheet'].copy(before=wb.sheets['Sheet1'])
print("คัดลอก 'RenamedSheet' ไว้ก่อน 'Sheet1' แล้ว")
```

```python
# 🔀 ย้าย worksheet — xlwings ไม่มีคำสั่งย้ายโดยตรง
# วิธีแก้: คัดลอก sheet ไปตำแหน่งใหม่ แล้วลบ sheet เดิม
# ขั้นตอน 1: คัดลอก 'Sheet1' ไว้ก่อน 'RenamedSheet'
# ขั้นตอน 2: ลบ 'Sheet1' เดิม
# ผลลัพธ์คือ "ย้าย" sheet ไปตำแหน่งใหม่
wb.sheets['Sheet1'].copy(before=wb.sheets['RenamedSheet'])
wb.sheets['Sheet1'].delete()
print("ย้าย 'Sheet1' แล้ว (คัดลอก + ลบ)")
```

## 8. ป้องกัน / ยกเลิกการป้องกัน Worksheet

```python
# 🔒 ป้องกัน worksheet — ป้องกันไม่ให้ผู้ใช้แก้ไขเนื้อหาเซลล์
# xlwings .api.Protect() ใช้ได้เฉพาะ Windows จึงใช้ openpyxl แทน
# openpyxl ทำงานได้ทุกแพลตฟอร์ม (macOS + Windows)
# ขั้นตอน: บันทึก → เปิดด้วย openpyxl → ป้องกัน → บันทึก → เปิดใหม่ด้วย xlwings
from openpyxl import load_workbook
file_path = wb.fullname      # ดึง path เต็มจาก xlwings
wb.save()                     # บันทึกการเปลี่ยนแปลงก่อน

ob = load_workbook(file_path) # เปิดไฟล์เดียวกันด้วย openpyxl
ob['Test3'].protection.sheet = True   # เปิดการป้องกัน sheet
ob['Test3'].protection.password = '1234'  # ตั้งรหัสผ่าน
ob.save(file_path)            # บันทึกด้วย openpyxl
ob.close()

wb = xw.Book(file_path)      # เปิดไฟล์ใหม่ใน xlwings
print("✅ Test3 ถูกป้องกันด้วยรหัสผ่าน '1234' แล้ว")
```

```python
# 🔓 ยกเลิกการป้องกัน worksheet — ลบการป้องกันเพื่อให้แก้ไขได้อีกครั้ง
# ใช้วิธีเดียวกัน: บันทึก → openpyxl → ยกเลิกป้องกัน → บันทึก → เปิดใหม่
file_path = wb.fullname
wb.save()

ob = load_workbook(file_path)
ob['Test3'].protection.sheet = False   # ปิดการป้องกัน
# ob['Test3'].protection.password = None  # ลบรหัสผ่าน
ob.save(file_path)
ob.close()

wb = xw.Book(file_path)      # เปิดใหม่ใน xlwings
print("✅ ยกเลิกการป้องกัน Test3 แล้ว")
```

## 9. ลบ Worksheet

```python
# 🗑️ ลบ worksheet — ลบ sheet ออกจาก workbook อย่างถาวร
# ⚠️ คำเตือน: การกระทำนี้ย้อนกลับไม่ได้!
# .delete() ทำงานข้ามแพลตฟอร์มผ่าน xlwings native method
# ระวังอย่าลบ sheet สุดท้าย — Excel ต้องมีอย่างน้อย 1 sheet
wb.sheets['Test2'].delete()
print("ลบ 'Test2' แล้ว")
```

## 10. บันทึกและปิด

```python
# 💾 บันทึกและปิด — ควรบันทึกงานก่อนปิดเสมอ!
# .save() เขียนการเปลี่ยนแปลงทั้งหมดลงไฟล์
# .close() ปิด workbook ในแอป Excel
# แนวปฏิบัติที่ดี: บันทึกก่อนปิดเสมอเพื่อไม่ให้ข้อมูลหาย
wb.save()
wb.close()
print("✅ บันทึกและปิด workbook สำเร็จ")
```
