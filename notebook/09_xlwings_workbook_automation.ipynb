```python
# 📚 นำเข้าไลบรารี
# xlwings — จัดการ Excel อัตโนมัติ (เปิด, อ่าน, เขียน, บันทึก) ด้วยการเชื่อมต่อ Excel จริง
# getpass — รับรหัสผ่านอย่างปลอดภัยโดยไม่แสดงบนหน้าจอ
import xlwings as xw
import getpass
```

## 1. สร้าง Excel App และ Workbook

```python
# 🚀 สร้าง Excel App — เปิดแอปพลิเคชัน Excel ขึ้นมาในพื้นหลัง
# visible=True คือแสดงหน้าต่าง Excel (ตั้ง False เพื่อทำงานแบบไม่แสดง)
# app.pid ให้ Process ID — มีประโยชน์เมื่อ Excel ค้าง
app = xw.App(visible=True)
print(f"สร้าง Excel App แล้ว PID: {app.pid}")
```

```python
# 📗 เพิ่ม workbook ใหม่ผ่าน app
# app.books.add() สร้าง workbook เปล่าใน Excel instance ที่กำลังทำงาน
# .name = แค่ชื่อไฟล์, .fullname = path เต็มบนดิสก์
wb = app.books.add()
print(f"ชื่อ Workbook: {wb.name}")
print(f"Path เต็ม: {wb.fullname}")
```

## 2. นับและนำทาง Workbook

```python
# 📊 นับ workbook ที่เปิดอยู่ — len(app.books) ให้จำนวนทั้งหมด
# วนลูปผ่าน app.books เพื่อแสดงรายการ workbook ที่เปิดอยู่
print(f"จำนวน workbook ที่เปิด: {len(app.books)}")

for book in app.books:
    print(f"  - {book.name}")
```

```python
# ➕ เพิ่ม workbook อีกอัน — สามารถเปิดหลาย workbook พร้อมกันได้
# app.books.add() แต่ละครั้งจะสร้าง workbook แยกกัน
wb2 = app.books.add()
print(f"Workbook ที่สอง: {wb2.name}")
print(f"จำนวน workbook ทั้งหมด: {len(app.books)}")
```

## 3. เปิด Workbook ที่มีอยู่แล้ว

```python
# 📂 เปิด workbook ที่มีอยู่จาก path ไฟล์
# xw.Book('path') เปิด workbook และเชื่อมต่อ
# read_only=True เปิดแบบอ่านอย่างเดียว (ปลอดภัยสำหรับการอ่านข้อมูล)

# ตัวอย่าง (uncomment แล้วเปลี่ยน path):
# wb_existing = xw.Book('path/to/your/file.xlsx')

# เปิดแบบอ่านอย่างเดียว:
# wb_readonly = app.books.open('path/to/file.xlsx', read_only=True)
```

## 4. นำทาง Workbook และ Worksheet

```python
# 🧭 นำทางระหว่าง workbook และ sheet
# app.books[0] = workbook แรก (index เริ่มจาก 0)
# first_wb.sheets วนลูปผ่าน sheet ทั้งหมดใน workbook นั้น
first_wb = app.books[0]
print(f"Workbook แรก: {first_wb.name}")

for sheet in first_wb.sheets:
    print(f"  Sheet: {sheet.name}")
```

## 5. บันทึก / บันทึกเป็น Workbook

```python
# ✍️ เขียนข้อมูลแล้วบันทึก workbook
# .range('A1').value = กำหนดค่าเซลล์เดียว
# .range('A2').value = [[...], [...]] เขียน list 2 มิติเป็นแถวข้อมูล
# .save('filename.xlsx') บันทึกไปยังไฟล์ที่ระบุ
wb.sheets[0].range('A1').value = [['app','cat','bat']]
wb.sheets[0].range('A2').value = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

wb.save('Workbook_Demo.xlsx')
print(f"บันทึกไปที่: {wb.fullname}")
```

```python
# 💾 บันทึกเป็น — บันทึก workbook เดิมด้วยชื่อ/ตำแหน่งใหม่
# os.path.join() สร้าง path เต็มอย่างปลอดภัย (ใช้ได้ทุก OS)
# มีประโยชน์เมื่อต้องการเก็บไฟล์เดิมและสร้างสำเนา
import os
save_path = os.path.join(os.getcwd(), 'Workbook_SaveAs_Demo.xlsx')
wb.save(save_path)
print(f"บันทึกเป็นไปที่: {save_path}")
```

## 6. ดู Path เต็มของ Workbook

```python
# 📍 ดูข้อมูล path ของ workbook
# .fullname = path เต็ม (เช่น /Users/.../Workbook_Demo.xlsx)
# .name = แค่ชื่อไฟล์ (เช่น Workbook_Demo.xlsx)
print(f"ชื่อเต็ม: {wb.fullname}")
print(f"ชื่อไฟล์อย่างเดียว: {wb.name}")
```

## 7. ~~ป้องกัน Workbook ด้วยรหัสผ่าน~~

```python
# 🔒 ป้องกันโครงสร้าง workbook ด้วยรหัสผ่าน
# xlwings .api.Protect() ใช้ได้เฉพาะ Windows (COM) จึงใช้ openpyxl แทนเพื่อรองรับทุกแพลตฟอร์ม
# ขั้นตอน: บันทึก → เปิดด้วย openpyxl → ตั้งรหัสผ่าน + ล็อคโครงสร้าง → บันทึก → เปิดใหม่
# getpass.getpass() ซ่อนรหัสผ่านขณะพิมพ์ (แนวปฏิบัติด้านความปลอดภัย)
import xlwings as xw
import openpyxl
from openpyxl.workbook.protection import WorkbookProtection
import getpass

# ขอรหัสผ่านจากผู้ใช้
password = getpass.getpass("กรุณาป้อนรหัสผ่านเพื่อป้องกันโครงสร้าง workbook: ")

save_path = "Workbook_Demo.xlsx"

# ลองบันทึก workbook xlwings ปัจจุบัน ถ้าไม่ได้จะสร้างไฟล์ตัวอย่างแทน
try:
    _ = wb.name  # ตรวจสอบว่า wb reference ยังใช้ได้
    wb.save()     # บันทึกก่อน (หลีกเลี่ยงข้อผิดพลาด save_as บน macOS)
    import shutil, os
    # คัดลอกไฟล์ที่บันทึกไปยัง path เป้าหมาย
    src = wb.fullname
    if os.path.abspath(src) != os.path.abspath(save_path):
        shutil.copy2(src, save_path)
    wb.close()
except Exception:
    # ถ้า xlwings ใช้ไม่ได้ สร้าง workbook ตัวอย่างด้วย openpyxl
    print("⚠️ xlwings workbook ไม่พร้อมใช้งาน — สร้างไฟล์ตัวอย่างเพื่อสาธิตการป้องกัน")
    temp_wb = openpyxl.Workbook()
    ws = temp_wb.active
    ws.title = "Sheet1"
    ws.append(["Name", "Value"])
    ws.append(["Sample", 100])
    temp_wb.save(save_path)
    temp_wb.close()

# โหลด workbook ด้วย openpyxl แล้วตั้งค่าป้องกัน
oxl_wb = openpyxl.load_workbook(save_path)
oxl_wb.security = WorkbookProtection(lockStructure=True)
oxl_wb.security.set_workbook_password(password)
oxl_wb.save(save_path)
oxl_wb.close()

# เปิดใหม่ด้วย xlwings
wb = xw.Book(save_path)
print("✅ ป้องกันโครงสร้าง workbook สำเร็จ!")
```

```python
# 🔓 ยกเลิกการป้องกัน workbook — ลบการป้องกันโครงสร้าง
# ใช้วิธีเดียวกัน: บันทึก → openpyxl → ล้างรหัสผ่าน → บันทึก → เปิดใหม่
import openpyxl

password = getpass.getpass("กรุณาป้อนรหัสผ่านเพื่อยกเลิกการป้องกัน: ")

# ใช้ path เดียวกับขั้นตอนป้องกัน
save_path = "Workbook_Demo.xlsx"
try:
    save_path = wb.fullname
    wb.save()
    wb.close()
except Exception:
    pass  # ไฟล์มีอยู่แล้วจากขั้นตอนป้องกัน

oxl_wb = openpyxl.load_workbook(save_path)
oxl_wb.security._workbook_password = ''      # ล้าง password hash โดยตรง
oxl_wb.security.lockStructure = False        # ปลดล็อคโครงสร้าง
oxl_wb.save(save_path)
oxl_wb.close()

wb = xw.Book(save_path)
print("✅ ยกเลิกการป้องกัน workbook สำเร็จ!")
```

## 8. เปิดหลาย Workbook พร้อมกัน

```python
# 📚 เปิดหลาย workbook ด้วยลูป
# glob.glob('*.xlsx') ค้นหาไฟล์ Excel ทั้งหมดในโฟลเดอร์ปัจจุบัน
# จำกัดไว้ 3 ไฟล์เพื่อไม่ให้เปิดมากเกินไป
# xw.Book(f) แต่ละครั้งจะเปิด workbook แยกกันใน Excel
import glob

excel_files = glob.glob('*.xlsx')
print(f"พบไฟล์ Excel {len(excel_files)} ไฟล์")

workbooks = []
for f in excel_files[:3]:  # จำกัดไว้ 3 ไฟล์
    wb_temp = xw.Book(f)
    workbooks.append(wb_temp)
    print(f"เปิดแล้ว: {wb_temp.name}")
```

## 9. ~~คุณสมบัติ Workbook~~

```python
# 📋 คุณสมบัติ workbook — ดูข้อมูลสรุปของ workbook ปัจจุบัน
# .sheets.active = sheet ที่เลือก/ใช้งานอยู่
# len(wb.sheets) = จำนวน sheet ทั้งหมด
# List comprehension [s.name for s in wb.sheets] = ชื่อ sheet ทั้งหมด
import xlwings as xw

# ลองใช้ wb reference เดิม ถ้าหมดอายุจะเชื่อมต่อใหม่
try:
    _ = wb.name
except Exception:
    print("⚠️ การอ้างอิง workbook หายไป กำลังเชื่อมต่อใหม่...")
    if xw.books:
        wb = xw.books.active
    else:
        wb = xw.Book()  # สร้าง workbook ใหม่ถ้าไม่มีเปิดอยู่
        print("📄 สร้าง workbook ใหม่เนื่องจากไม่มี workbook เปิดอยู่")

print(f"Workbook: {wb.name}")
print(f"รายชื่อ Sheet: {[s.name for s in wb.sheets]}")
print(f"จำนวน Sheet ทั้งหมด: {len(wb.sheets)}")
```
