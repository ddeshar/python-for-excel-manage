รันเซลล์ด้านล่างเพื่อติดตั้ง package ทั้งหมดที่ใช้ใน Notebook:

| Package | ใช้ใน Notebook | ใช้ทำอะไร |
|---------|---------------|----------|

| python-dotenv | 15-17 | โหลดค่ารหัสผ่านจากไฟล์ .env |

````python
# 📦 ติดตั้ง package ทั้งหมดอัตโนมัติ (กดรันครั้งเดียว)
# ระบบจะตรวจสอบแต่ละ package — ติดตั้งเฉพาะตัวที่ยังไม่มี
import subprocess, sys

REQUIRED_PACKAGES = {
    'openpyxl':        {'ใช้ใน': 'NB 02-06', 'ทำอะไร': 'อ่าน/เขียน Excel แบบไม่ต้องเปิด Excel'},
    'pandas':          {'ใช้ใน': 'NB 06-08, 11-17', 'ทำอะไร': 'จัดการข้อมูลตาราง'},
    'numpy':           {'ใช้ใน': 'NB 06', 'ทำอะไร': 'คำนวณตัวเลข'},
    'xlwings':         {'ใช้ใน': 'NB 09-13', 'ทำอะไร': 'ควบคุม Excel แบบเปิดจริง'},
    'requests':        {'ใช้ใน': 'NB 04-05, 11-12', 'ทำอะไร': 'ดาวน์โหลดไฟล์จากอินเทอร์เน็ต'},
    'matplotlib':      {'ใช้ใน': 'NB 06', 'ทำอะไร': 'สร้างกราฟ'},
    'sqlalchemy':      {'ใช้ใน': 'NB 15-17', 'ทำอะไร': 'เชื่อมต่อ Python กับ Database'},
    'psycopg2-binary': {'ใช้ใน': 'NB 15-17', 'ทำอะไร': 'Driver สำหรับ PostgreSQL'},
    'python-dotenv':   {'ใช้ใน': 'NB 15-17', 'ทำอะไร': 'โหลดค่ารหัสผ่านจากไฟล์ .env'},
}

installed_count = 0
installed_new = 0
failed = []

print("=" * 60)
print("🔧 ตรวจสอบและติดตั้ง Python Packages")
print("=" * 60)

for pkg, info in REQUIRED_PACKAGES.items():
    # psycopg2-binary import ด้วยชื่อ psycopg2, python-dotenv import ด้วยชื่อ dotenv
    import_name = {'psycopg2-binary': 'psycopg2', 'python-dotenv': 'dotenv'}.get(pkg, pkg)
    try:
        mod = __import__(import_name)
        ver = getattr(mod, '__version__', '?')
        print(f"  ✅ {pkg} ({ver}) — มีอยู่แล้ว")
        installed_count += 1
    except ImportError:
        print(f"  ⏳ {pkg} — ยังไม่มี กำลังติดตั้ง...")
        try:
            subprocess.check_call(
                [sys.executable, '-m', 'pip', 'install', pkg, '-q'],
                stdout=subprocess.DEVNULL
            )
            mod = __import__(import_name)
            ver = getattr(mod, '__version__', '?')
            print(f"  ✅ {pkg} ({ver}) — ติดตั้งสำเร็จ!")
            installed_count += 1
            installed_new += 1
        except Exception as e:
            print(f"  ❌ {pkg} — ติดตั้งไม่สำเร็จ: {e}")
            failed.append(pkg)

print("=" * 60)
if failed:
    print(f"⚠️ ติดตั้งไม่สำเร็จ {len(failed)} package: {', '.join(failed)}")
    print("💡 ลองรันเซลล์นี้อีกครั้ง หรือเปิด Terminal แล้วพิมพ์:")
    for pkg in failed:
        print(f"   pip install {pkg}")
else:
    print(f"🎉 ติดตั้งครบทั้ง {installed_count} packages!")
    if installed_new > 0:
        print(f"   (ติดตั้งใหม่ {installed_new} ตัว)")
    else:
        print("   (มีครบทุกตัวอยู่แล้ว — ไม่ต้องทำอะไรเพิ่ม)")
print("=" * 60)
````

> 💡 **สำคัญ:** รันเซลล์ที่ 1/6 ก่อนเสมอ (เพราะเป็นตัวสร้างตัวแปร `alerts` ที่ใช้ร่วมกัน)

````python
# 🩺 (1/6) ตรวจ Python, OS, pip
import sys, platform

alerts = []  # เก็บปัญหาที่พบ — ใช้ร่วมกันทุกเซลล์ (รันเซลล์นี้ก่อนเสมอ)

print("=" * 60)
print("🩺 ตรวจสอบ Python / OS / pip")
print("=" * 60)

# Python
py_ver = sys.version.split()[0]
py_major, py_minor = sys.version_info[:2]
if py_major >= 3 and py_minor >= 10:
    print(f"  ✅ Python: {py_ver}")
else:
    print(f"  ⚠️ Python: {py_ver} (แนะนำ 3.10 ขึ้นไป)")
    alerts.append(f"Python เวอร์ชัน {py_ver} เก่าเกินไป — แนะนำ Python 3.10+")

# OS
os_name = platform.system()
os_detail = platform.platform()
if os_name == 'Windows':
    print(f"  ✅ ระบบปฏิบัติการ: Windows ({os_detail})")
elif os_name == 'Darwin':
    print(f"  ℹ️ ระบบปฏิบัติการ: macOS ({os_detail})")
    print(f"     (Notebook นี้ออกแบบสำหรับ Windows แต่ใช้กับ macOS ได้เช่นกัน)")
else:
    print(f"  ℹ️ ระบบปฏิบัติการ: {os_name} ({os_detail})")

# pip
try:
    import pip
    print(f"  ✅ pip: พร้อมใช้งาน")
except ImportError:
    print(f"  ❌ pip: ไม่พบ")
    alerts.append("pip ไม่พร้อมใช้งาน — ลองรัน: python -m ensurepip --upgrade")

print("=" * 60)
````

````python
# 🩺 (3/6) ตรวจ Microsoft Excel
import os, platform

print("=" * 60)
print("📊 สถานะ Excel")
print("=" * 60)

os_name = platform.system()
excel_found = False

if os_name == 'Windows':
    excel_paths = [
        os.path.expandvars(r'%ProgramFiles%\Microsoft Office'),
        os.path.expandvars(r'%ProgramFiles(x86)%\Microsoft Office'),
        os.path.expandvars(r'%LOCALAPPDATA%\Microsoft\WindowsApps'),
    ]
    for path in excel_paths:
        if os.path.exists(path):
            excel_found = True
            break
    try:
        import xlwings as xw
        app = xw.App(visible=False)
        app.quit()
        excel_found = True
        print(f"  ✅ Excel: พร้อมใช้งานกับ xlwings (NB 09-13)")
    except Exception:
        if excel_found:
            print(f"  ⚠️ Excel: พบ Microsoft Office แต่ยังไม่ได้ทดสอบ xlwings")
        else:
            print(f"  ⚠️ Excel: ไม่พบ Microsoft Excel")
            alerts.append("ไม่พบ Microsoft Excel — จำเป็นสำหรับ NB 09-13 (xlwings)\n     💡 NB 01-08 ใช้ได้โดยไม่ต้องมี Excel")
elif os_name == 'Darwin':
    if os.path.exists('/Applications/Microsoft Excel.app'):
        print(f"  ✅ Excel: พบ Microsoft Excel สำหรับ Mac")
    else:
        print(f"  ⚠️ Excel: ไม่พบ Microsoft Excel สำหรับ Mac")
        alerts.append("ไม่พบ Microsoft Excel — จำเป็นสำหรับ NB 09-13 (xlwings)\n     💡 NB 01-08 ใช้ได้โดยไม่ต้องมี Excel")

print("=" * 60)
````

## ขั้นตอนที่ 5: ดาวน์โหลดไฟล์ข้อมูลตัวอย่าง

รันเซลล์ด้านล่างเพื่อดาวน์โหลดไฟล์ที่ใช้ใน Notebook 04-06, 11-12 อัตโนมัติ

| ไฟล์ | ใช้ใน | ข้อมูล |
|------|-------|--------|
| Random Sales Data.xlsx | NB 04-05, 11-12 | ข้อมูลยอดขาย (Sales Representative, Region, Product ฯลฯ) |
| superstore.xlsx | NB 06 | ข้อมูล Superstore (Order ID, Customer, Category, Sales ฯลฯ) |

````python
# 📥 ดาวน์โหลดไฟล์ข้อมูลตัวอย่างอัตโนมัติ
# ตรวจสอบว่ามีไฟล์หรือยัง — ถ้าไม่มีจะดาวน์โหลดให้
import os, requests

DATA_FILES = {
    'Random Sales Data.xlsx': {
        'url': 'https://raw.githubusercontent.com/ddeshar/python-for-excel-manage/main/Random%20Sales%20Data.xlsx',
        'ใช้ใน': 'NB 04-05, 11-12',
    },
    'superstore.xlsx': {
        'url': 'https://raw.githubusercontent.com/ddeshar/python-for-excel-manage/main/Superstore.xlsx',
        'ใช้ใน': 'NB 06',
    },
}

print("=" * 60)
print("📥 ตรวจสอบและดาวน์โหลดไฟล์ข้อมูลตัวอย่าง")
print("=" * 60)

downloaded = 0
skipped = 0
failed_files = []

for filename, info in DATA_FILES.items():
    if os.path.exists(filename):
        size_kb = os.path.getsize(filename) / 1024
        print(f"  ✅ {filename} ({size_kb:.0f} KB) — มีอยู่แล้ว")
        skipped += 1
    else:
        print(f"  ⏳ กำลังดาวน์โหลด {filename}...")
        try:
            response = requests.get(info['url'], timeout=30)
            response.raise_for_status()
            with open(filename, 'wb') as f:
                f.write(response.content)
            size_kb = len(response.content) / 1024
            print(f"  ✅ {filename} ({size_kb:.0f} KB) — ดาวน์โหลดสำเร็จ!")
            downloaded += 1
        except Exception as e:
            print(f"  ❌ {filename} — ดาวน์โหลดไม่สำเร็จ: {e}")
            failed_files.append(filename)

print("=" * 60)
if failed_files:
    print(f"⚠️ ดาวน์โหลดไม่สำเร็จ {len(failed_files)} ไฟล์:")
    for f in failed_files:
        print(f"   ❌ {f}")
    print("💡 ตรวจสอบการเชื่อมต่ออินเทอร์เน็ต แล้วรันเซลล์นี้อีกครั้ง")
else:
    total = downloaded + skipped
    print(f"🎉 ไฟล์ข้อมูลพร้อมครบ {total} ไฟล์!")
    if downloaded > 0:
        print(f"   (ดาวน์โหลดใหม่ {downloaded} ไฟล์)")
print("=" * 60)
````

## ขั้นตอนที่ 7: ติดตั้ง Docker Desktop (สำหรับ NB 15-17)

Notebook 15-17 ใช้ **PostgreSQL** เป็นฐานข้อมูล โดยรันผ่าน **Docker** (ไม่ต้องติดตั้ง PostgreSQL เอง)

### Docker คืออะไร?
Docker = โปรแกรมที่ช่วยรัน "คอนเทนเนอร์" (เหมือนกล่องสำเร็จรูป) — เราใช้มันรัน PostgreSQL โดยไม่ต้องติดตั้งลงเครื่องจริง

### วิธีติดตั้ง

#### Windows
1. ไปที่ [docker.com/products/docker-desktop](https://www.docker.com/products/docker-desktop/)
2. กด **Download for Windows**
3. รันตัวติดตั้ง (Docker Desktop Installer.exe)
4. ✅ ติ๊กช่อง **"Use WSL 2 instead of Hyper-V"** (แนะนำ)
5. กด **OK** → รอติดตั้ง → **Close and restart** (ต้องรีสตาร์ทเครื่อง)
6. หลังรีสตาร์ท เปิด **Docker Desktop** จาก Start Menu
7. ยอมรับเงื่อนไข → รอจน Docker พร้อม (ไอคอนปลาวาฬสีเขียวที่ taskbar)

> ⚠️ **Windows 10/11 Home**: ต้องเปิด WSL 2 ก่อน — เปิด PowerShell (Admin) แล้วพิมพ์:
> ```
> wsl --install
> ```
> รีสตาร์ทเครื่อง แล้วค่อยติดตั้ง Docker Desktop

#### macOS
1. ไปที่ [docker.com/products/docker-desktop](https://www.docker.com/products/docker-desktop/)
2. กด **Download for Mac** (เลือก Apple Chip หรือ Intel ตามเครื่อง)
3. เปิดไฟล์ .dmg → ลาก Docker ไปใส่ Applications
4. เปิด Docker จาก Applications → ยอมรับเงื่อนไข → รอจนพร้อม

### ตรวจสอบว่าติดตั้งสำเร็จ
เปิด **Terminal** (macOS) หรือ **Command Prompt** (Windows) แล้วพิมพ์:
```
docker --version
docker compose version
```
ถ้าขึ้นเลขเวอร์ชัน แสดงว่าสำเร็จ ✅

### รัน PostgreSQL ด้วย Docker
ในโฟลเดอร์โปรเจกต์นี้มีไฟล์ `docker-compose.yml` อยู่แล้ว — รันคำสั่งนี้ใน Terminal:
```
docker compose up -d
```
- `-d` = รันเบื้องหลัง (ไม่ต้องเปิด Terminal ค้างไว้)
- หยุด: `docker compose down`

> 💡 NB 01-14 **ไม่ต้องใช้ Docker** — เรียนได้เลยทันที!

````python
# 🐳 ตรวจสอบ Docker — รันเซลล์นี้เพื่อเช็คว่า Docker พร้อมใช้งาน
import subprocess, shutil

print("=" * 60)
print("🐳 ตรวจสอบ Docker Desktop")
print("=" * 60)

# ตรวจว่ามี docker คำสั่งหรือไม่
docker_path = shutil.which("docker")
if not docker_path:
    print("  ❌ ไม่พบคำสั่ง docker — กรุณาติดตั้ง Docker Desktop ก่อน")
    print("     💡 ดาวน์โหลดที่: https://www.docker.com/products/docker-desktop/")
else:
    # ตรวจเวอร์ชัน docker
    try:
        result = subprocess.run(["docker", "--version"], capture_output=True, text=True, timeout=10)
        print(f"  ✅ {result.stdout.strip()}")
    except Exception as e:
        print(f"  ⚠️ docker พบแต่รันไม่ได้: {e}")

    # ตรวจ docker compose
    try:
        result = subprocess.run(["docker", "compose", "version"], capture_output=True, text=True, timeout=10)
        print(f"  ✅ {result.stdout.strip()}")
    except Exception:
        print("  ⚠️ docker compose ไม่พร้อม")

    # ตรวจว่า Docker daemon ทำงานอยู่หรือไม่
    try:
        result = subprocess.run(["docker", "info"], capture_output=True, text=True, timeout=10)
        if result.returncode == 0:
            print("  ✅ Docker Desktop กำลังทำงาน")
        else:
            print("  ⚠️ Docker Desktop ยังไม่ได้เปิด — เปิด Docker Desktop ก่อนรัน NB 15-17")
    except Exception:
        print("  ⚠️ ตรวจสอบ Docker ไม่ได้ — เปิด Docker Desktop แล้วลองอีกครั้ง")

    # ตรวจว่า PostgreSQL container รันอยู่หรือไม่
    try:
        result = subprocess.run(
            ["docker", "compose", "ps", "--format", "{{.Name}} {{.Status}}"],
            capture_output=True, text=True, timeout=10
        )
        if result.returncode == 0 and result.stdout.strip():
            print(f"\n  📦 Containers ที่กำลังรัน:")
            for line in result.stdout.strip().split("\n"):
                print(f"     {line}")
        else:
            print(f"\n  ℹ️ ยังไม่มี container รัน — รันคำสั่งนี้ใน Terminal:")
            print(f"     docker compose up -d")
    except Exception:
        pass

print("=" * 60)
````

````python
# ✅ Checklist (1/3) — ตรวจ Python + Packages
import sys

print("=" * 60)
print("✅ CHECKLIST — ตรวจสอบความพร้อมก่อนเริ่มเรียน")
print("=" * 60)

all_ok = True

# Python
py_ver = sys.version_info
check = py_ver.major >= 3 and py_ver.minor >= 10
status = "✅" if check else "❌"
if not check: all_ok = False
print(f"  {status} Python 3.10+ — (ปัจจุบัน: {py_ver.major}.{py_ver.minor}.{py_ver.micro})")
if not check:
    print(f"     ⚠️ กรุณาอัพเดต Python จาก python.org")

# Packages (รวม DB packages ด้วย)
print()
pkgs = {
    'openpyxl': 'openpyxl',
    'pandas': 'pandas',
    'numpy': 'numpy',
    'xlwings': 'xlwings',
    'requests': 'requests',
    'matplotlib': 'matplotlib',
    'sqlalchemy': 'sqlalchemy',
    'psycopg2': 'psycopg2',       # ติดตั้งชื่อ psycopg2-binary
    'dotenv': 'python-dotenv',     # ติดตั้งชื่อ python-dotenv
}
missing_pkgs = []
for import_name, install_name in pkgs.items():
    try:
        __import__(import_name)
        print(f"  ✅ {install_name}")
    except ImportError:
        print(f"  ❌ {install_name} — ยังไม่ได้ติดตั้ง!")
        missing_pkgs.append(install_name)
        all_ok = False

if missing_pkgs:
    print(f"\n  ⚠️ Package ที่ขาด: {', '.join(missing_pkgs)}")
    print(f"  💡 วิธีแก้: รันเซลล์ติดตั้งด้านบนอีกครั้ง (ขั้นตอนที่ 4)")
````
