````python
# === Lab 3 เฉลย ===










df = pd.read_excel("output/labs/icp_samples.xlsx")

# 1. แปลงเป็น list of dict
records = df.to_dict(orient="records")
# แปลง Timestamp เป็น string
for r in records:
    if "test_date" in r and hasattr(r["test_date"], "isoformat"):
        r["test_date"] = str(r["test_date"])

# 2. สร้าง JSON structure
report = {
    "report_date": str(pd.Timestamp.now().date()),
    "total_samples": len(df),
    "summary": {
        "by_lab": df["lab"].value_counts().to_dict(),
        "by_status": df["status"].value_counts().to_dict(),
    },
    "samples": records
}

# 3. บันทึก
with open("output/labs/report.json", "w", encoding="utf-8") as f:
    json.dump(report, f, indent=2, ensure_ascii=False, default=str)
print("✅ Saved report.json")

# 4. อ่านกลับ
with open("output/labs/report.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)

print(f"Report date: {loaded['report_date']}")
print(f"Total: {loaded['total_samples']}")
print(f"By lab: {loaded['summary']['by_lab']}")

df_back = pd.DataFrame(loaded["samples"])
print(f"\nDataFrame shape: {df_back.shape}")
df_back.head(3)
````

2. Insert ข้อมูล 5 แถวจาก DataFrame (ใช้ `to_sql`)

````python
# Lab 4: YOUR CODE HERE

# 1. สร้างตาราง lab4_samples
# with engine.connect() as conn: ...

# 2. Insert 5 แถว
# df_5 = ...
# df_5.to_sql(...)

# 3. SELECT ทั้งหมด
# pd.read_sql(...)

# 4. UPDATE
# with engine.connect() as conn: ...

# 5. DELETE
# with engine.connect() as conn: ...
````

````python
# === Lab 4 เฉลย ===










# 1. สร้างตาราง
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS lab4_samples CASCADE"))
    conn.execute(text("""
        CREATE TABLE lab4_samples (
            id SERIAL PRIMARY KEY,
            sample_id TEXT,
            analyst TEXT,
            lab TEXT,
            As_value REAL,
            Pb_value REAL,
            status TEXT
        )
    """))
    conn.commit()
print("✅ Table created")

# 2. Insert 5 แถว
df = pd.read_excel("output/labs/icp_samples.xlsx")
df_5 = df.head(5)[["sample_id", "analyst", "lab", "As_mgkg", "Pb_mgkg", "status"]]
df_5.columns = ["sample_id", "analyst", "lab", "As_value", "Pb_value", "status"]
df_5.to_sql("lab4_samples", engine, if_exists="append", index=False)
print(f"✅ Inserted {len(df_5)} rows")

# 3. SELECT
result = pd.read_sql("SELECT * FROM lab4_samples", engine)
print("\n📊 All samples:")
print(result.to_string())

# 4. UPDATE
with engine.connect() as conn:
    conn.execute(text(
        "UPDATE lab4_samples SET status = 'approved' WHERE sample_id = 'ICP-0001'"
    ))
    conn.commit()
print("\n✅ Updated ICP-0001 to approved")

# 5. DELETE
with engine.connect() as conn:
    conn.execute(text("DELETE FROM lab4_samples WHERE status = 'failed'"))
    conn.commit()
print(f"✅ Deleted failed samples")

# ตรวจสอบ
final = pd.read_sql("SELECT * FROM lab4_samples", engine)
print(f"\n📊 Final ({len(final)} rows):")
print(final.to_string())
````

## ⭐⭐ Lab 5: SQL Queries

### โจทย์: ใช้ database ที่สร้างไว้
1. สร้าง database ใหม่ พร้อมข้อมูล 30 rows
2. เขียน SQL: หาจำนวน samples ตาม lab
3. เขียน SQL: หาค่าเฉลี่ย, min, max ของ As แยกตาม analyst
4. เขียน SQL: หา samples ที่ As > ค่าเฉลี่ย (ใช้ Subquery)
5. เขียน SQL: Rank samples ตาม Pb จากมากไปน้อย (ใช้ Window Function)

````python
# Lab 5: Setup — โหลดข้อมูล 30 rows เข้า PostgreSQL
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS lab5_samples CASCADE"))
    conn.commit()

df = pd.read_excel("output/labs/icp_samples.xlsx")
df_db = df[["sample_id", "analyst", "lab", "panel", "As_mgkg", "Pb_mgkg", "Cd_mgkg", "Hg_mgkg", "status"]]
df_db.to_sql("lab5_samples", engine, if_exists="replace", index=False)
print(f"✅ Loaded {len(df_db)} rows to PostgreSQL")
````

````python
# Lab 5: YOUR CODE HERE

# 2. จำนวน samples ตาม lab
# pd.read_sql("SELECT ...", engine)

# 3. ค่าเฉลี่ย, min, max ของ As ตาม analyst
# pd.read_sql("SELECT ...", engine)

# 4. As > ค่าเฉลี่ย (Subquery)
# pd.read_sql("SELECT ... WHERE \"As_mgkg\" > (SELECT ...)", engine)

# 5. Rank ตาม Pb (Window Function)
# pd.read_sql("SELECT ..., ROW_NUMBER() OVER (...) ...", engine)
````

````python
# === Lab 5 เฉลย ===










# 2. จำนวน samples ตาม lab
print("=== 2. Samples per lab ===")
q2 = pd.read_sql("""
    SELECT lab, COUNT(*) AS sample_count
    FROM lab5_samples
    GROUP BY lab
    ORDER BY sample_count DESC
""", engine)
print(q2.to_string())

# 3. As stats ตาม analyst
print("\n=== 3. As statistics per analyst ===")
q3 = pd.read_sql("""
    SELECT 
        analyst,
        COUNT(*) AS count,
        ROUND(AVG("As_mgkg")::numeric, 3) AS avg_As,
        ROUND(MIN("As_mgkg")::numeric, 3) AS min_As,
        ROUND(MAX("As_mgkg")::numeric, 3) AS max_As
    FROM lab5_samples
    GROUP BY analyst
""", engine)
print(q3.to_string())

# 4. As > average (Subquery)
print("\n=== 4. Samples with above-average As ===")
q4 = pd.read_sql("""
    SELECT sample_id, analyst, "As_mgkg"
    FROM lab5_samples
    WHERE "As_mgkg" > (SELECT AVG("As_mgkg") FROM lab5_samples)
    ORDER BY "As_mgkg" DESC
""", engine)
print(f"Found {len(q4)} samples")
print(q4.head(5).to_string())

# 5. Rank by Pb (Window Function)
print("\n=== 5. Samples ranked by Pb ===")
q5 = pd.read_sql("""
    SELECT 
        sample_id,
        lab,
        "Pb_mgkg",
        ROW_NUMBER() OVER (ORDER BY "Pb_mgkg" DESC) AS overall_rank,
        ROW_NUMBER() OVER (PARTITION BY lab ORDER BY "Pb_mgkg" DESC) AS lab_rank
    FROM lab5_samples
    ORDER BY overall_rank
    LIMIT 10
""", engine)
print(q5.to_string())
````

## ⭐⭐⭐ Lab 6: Mini ETL Pipeline

### โจทย์: สร้าง ETL Pipeline แบบง่าย

1. **Extract**: อ่าน `icp_samples.xlsx`
2. **Transform**:
   - กรองเฉพาะ status = 'completed'
   - เพิ่มคอลัมน์ `total_heavy_metals` = As + Pb + Cd + Hg
   - เพิ่มคอลัมน์ `risk_level`:
     - 'HIGH' ถ้า total > 10
     - 'MEDIUM' ถ้า total > 5
     - 'LOW' ถ้า total <= 5
   - แปลง test_date เป็น string
3. **Load**:
   - บันทึกลง PostgreSQL ตาราง `lab6_clean_results`
   - บันทึก JSON summary เป็น `output/labs/etl_summary.json`
4. **Verify**: SELECT จาก database เพื่อตรวจสอบ

````python
# Lab 6: เขียนโค้ดของคุณที่นี่

# 1. Extract (ดึงข้อมูล)
# df = ...

# 2. Transform (แปลงข้อมูล)
# df_clean = ...
# df_clean["total_heavy_metals"] = ...  # คำนวณผลรวมโลหะหนัก
# df_clean["risk_level"] = ...          # กำหนดระดับความเสี่ยง

# 3. Load to PostgreSQL (เขียนลง Database)
# df_clean.to_sql("lab6_clean_results", engine, ...)

# 3b. Save JSON summary (บันทึกสรุปเป็น JSON)
# summary = { ... }

# 4. Verify (ตรวจสอบผล)
# pd.read_sql("SELECT ...", engine)
````

````python
# === Lab 6 เฉลย ===










print("=" * 60)
print("🔄 ETL Pipeline - Lab 6")
print("=" * 60)

# 1. Extract
print("\n📥 Step 1: Extract")
df = pd.read_excel("output/labs/icp_samples.xlsx")
print(f"   Read {len(df)} rows")

# 2. Transform
print("\n🔧 Step 2: Transform")
df_clean = df[df["status"] == "completed"].copy()
print(f"   Filtered to {len(df_clean)} completed samples")

# เพิ่ม total
df_clean["total_heavy_metals"] = (
    df_clean["As_mgkg"] + df_clean["Pb_mgkg"] + 
    df_clean["Cd_mgkg"] + df_clean["Hg_mgkg"]
).round(3)

# เพิ่ม risk level
def get_risk(total):
    if total > 10:
        return "HIGH"
    elif total > 5:
        return "MEDIUM"
    return "LOW"

df_clean["risk_level"] = df_clean["total_heavy_metals"].apply(get_risk)

# แปลง date
df_clean["test_date"] = df_clean["test_date"].astype(str)

print(f"   Risk distribution:")
print(f"   {df_clean['risk_level'].value_counts().to_dict()}")

# 3. Load
print("\n📤 Step 3: Load")
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS lab6_clean_results CASCADE"))
    conn.commit()
df_clean.to_sql("lab6_clean_results", engine, if_exists="replace", index=False)
print(f"   Saved {len(df_clean)} rows to PostgreSQL")

# JSON summary
summary = {
    "pipeline": "Lab 6 Mini ETL",
    "run_at": pd.Timestamp.now().isoformat(),
    "input_rows": len(df),
    "output_rows": len(df_clean),
    "filtered_out": len(df) - len(df_clean),
    "risk_distribution": df_clean["risk_level"].value_counts().to_dict(),
    "avg_total_metals": round(df_clean["total_heavy_metals"].mean(), 3),
}

with open("output/labs/etl_summary.json", "w") as f:
    json.dump(summary, f, indent=2)
print(f"   Saved summary to etl_summary.json")

# 4. Verify
print("\n✅ Step 4: Verify")
verify = pd.read_sql("""
    SELECT 
        risk_level,
        COUNT(*) AS count,
        ROUND(AVG(total_heavy_metals)::numeric, 3) AS avg_total
    FROM lab6_clean_results
    GROUP BY risk_level
    ORDER BY avg_total DESC
""", engine)
print(verify.to_string())

print(f"\n{'='*60}")
print("✅ ETL Pipeline Complete!")
````
