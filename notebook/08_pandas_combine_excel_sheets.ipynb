## ขั้นตอนที่ 1: ตรวจสอบไฟล์ Excel ต้นทาง
โหลดและตรวจสอบไฟล์ Excel ที่มีหลาย sheet

```python
# 📥 ดาวน์โหลดไฟล์ข้อมูลจาก GitHub (ถ้ายังไม่มี)
# ไฟล์นี้มีข้อมูลแยกอยู่ในหลาย sheet — เหมาะสำหรับฝึกรวมข้อมูล
import requests
import os

url = "https://raw.githubusercontent.com/AbhisheakSaraswat/RawData/main/Splited%20Raw%20Data.xlsx"
file_path = "Splited Raw Data.xlsx"

if os.path.exists(file_path):
    print(f"✅ {file_path} มีอยู่แล้ว ข้ามการดาวน์โหลด")
else:
    print("📥 ไม่พบไฟล์ กำลังดาวน์โหลด...")
    r = requests.get(url)
    r.raise_for_status()  # แสดง error ถ้าดาวน์โหลดไม่สำเร็จ
    with open(file_path, "wb") as f:
        f.write(r.content)
    print("✅ ดาวน์โหลดไฟล์สำเร็จ")

# 📋 โหลดไฟล์ Excel แล้วตรวจสอบว่ามี sheet อะไรบ้าง
xls = pd.ExcelFile(file_path)
print(f"\nSheet ที่มีอยู่: {xls.sheet_names}")
print(f"จำนวน sheet: {len(xls.sheet_names)}")
```

## ขั้นตอนที่ 2: อ่านทุก Sheet เป็น DataFrame
อ่านแต่ละ sheet แล้วแสดงตัวอย่าง

```python
# 📖 อ่านทุก sheet ในครั้งเดียวเป็น dictionary: {ชื่อ_sheet: DataFrame}
# sheet_name=None บอก pandas ให้อ่านทุก sheet
all_sheets = pd.read_excel(file_path, sheet_name=None)

# 🔍 แสดงสรุปแต่ละ sheet (ชื่อ, จำนวนแถว, จำนวนคอลัมน์)
for sheet_name, df in all_sheets.items():
    print(f"Sheet: '{sheet_name}' -> {df.shape[0]} แถว, {df.shape[1]} คอลัมน์")
```

## ขั้นตอนที่ 3: รวมทุก Sheet เป็น DataFrame เดียว
ใช้ `pd.concat()` เพื่อรวมทุก sheet

```python
# 🔗 รวมทุก sheet เป็น DataFrame เดียวด้วย pd.concat()
# .values() ให้ list ของ DataFrame (หนึ่งตัวต่อ sheet)
# ignore_index=True รีเซ็ตเลขแถว (0, 1, 2...) แทนที่จะเก็บ index เดิม
combined_df = pd.concat(all_sheets.values(), ignore_index=True)

print(f"ขนาด DataFrame รวม: {combined_df.shape}")
combined_df.head()
```

## ขั้นตอนที่ 4: บันทึกข้อมูลที่รวมแล้ว
ส่งออก DataFrame รวมเป็นไฟล์ Excel ใหม่

```python
# 💾 บันทึกข้อมูลรวมเป็นไฟล์ Excel ใหม่
# engine='openpyxl' ถูกใช้โดยอัตโนมัติสำหรับไฟล์ .xlsx
output_file = "Combined_Data.xlsx"
combined_df.to_excel(output_file, index=False, sheet_name='Combined')
print(f"✅ บันทึกข้อมูลรวมไปที่: {output_file}")
print(f"จำนวนแถวรวม: {len(combined_df)}")
```

## ทางเลือก: รวม Sheet พร้อมเพิ่มคอลัมน์ชื่อ Sheet ต้นทาง
เพิ่มคอลัมน์ระบุว่าแต่ละแถวมาจาก sheet ไหน

```python
# 🏷️ รวมพร้อมคอลัมน์ "Source_Sheet" เพื่อรู้ว่าแต่ละแถวมาจาก sheet ไหน
# มีประโยชน์สำหรับการตรวจสอบย้อนกลับ — สามารถติดตามข้อมูลกลับไปที่ต้นทางได้
dfs_with_source = []
for sheet_name, df in all_sheets.items():
    df['Source_Sheet'] = sheet_name  # เพิ่มคอลัมน์ชื่อ sheet ต้นทาง
    dfs_with_source.append(df)

combined_with_source = pd.concat(dfs_with_source, ignore_index=True)
print(f"ข้อมูลรวมพร้อมคอลัมน์ต้นทาง: {combined_with_source.shape}")
combined_with_source.head()
```

## รวมหลายไฟล์ Excel จากโฟลเดอร์
โบนัส: ถ้าคุณมีหลายไฟล์ Excel ในโฟลเดอร์เดียวกัน

```python
# 📁 โบนัส: รวมหลายไฟล์ Excel จากโฟลเดอร์ (ไม่ใช่แค่ sheet แต่เป็นทั้งไฟล์)
# มีประโยชน์เมื่อได้รับหลายรายงาน (เช่น ม.ค..xlsx, ก.พ..xlsx ฯลฯ)
import glob
import os

folder_path = "./data/"  # 📂 เปลี่ยน path นี้เป็นโฟลเดอร์ของคุณ

# 🔍 ค้นหาไฟล์ .xlsx ทั้งหมดในโฟลเดอร์
excel_files = glob.glob(os.path.join(folder_path, "*.xlsx"))
print(f"พบ {len(excel_files)} ไฟล์ Excel")

if excel_files:
    all_dfs = []
    for file in excel_files:
        df = pd.read_excel(file)
        df['Source_File'] = os.path.basename(file)  # ติดตามว่าแต่ละแถวมาจากไฟล์ไหน
        all_dfs.append(df)
    
    combined = pd.concat(all_dfs, ignore_index=True)
    print(f"จำนวนแถวรวมทั้งหมด: {len(combined)}")

    output_file = os.path.join(folder_path, "combined_output.xlsx")
    combined.to_excel(output_file, index=False)

    print(f"✅ บันทึกไฟล์รวม: {output_file}")
```
