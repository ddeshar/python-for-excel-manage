## เตรียมตัว: เชื่อมต่อ Database

รันเซลล์นี้ก่อนเพื่อเชื่อมต่อ Database (เหมือนที่ทำในบทเรียน)

```python
# === เชื่อมต่อ Database (คัดลอกจาก NB 15) ===
import os
import pandas as pd  # pd = ชื่อย่อของ pandas ที่นิยมใช้กัน
from dotenv import load_dotenv  # สำหรับโหลดค่าจากไฟล์ .env
from sqlalchemy import create_engine, text  # สำหรับเชื่อมต่อ Database

# โหลดค่าจากไฟล์ .env (เก็บรหัสผ่านไว้ที่นี่)
load_dotenv('../.env')

# สร้าง Database URL จากค่าใน .env
DB_HOST = os.getenv('DB_HOST', 'localhost')
DB_PORT = os.getenv('DB_PORT', '5432')
DB_NAME = os.getenv('DB_NAME', 'icp_training')
DB_USER = os.getenv('DB_USER', 'icp_user')
DB_PASSWORD = os.getenv('DB_PASSWORD', 'training_password')
DATABASE_URL = f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'

# engine = ตัวเชื่อมต่อไปยัง Database
engine = create_engine(DATABASE_URL, echo=False)
print(f'✅ เชื่อมต่อสำเร็จ: {engine.url}')
```
