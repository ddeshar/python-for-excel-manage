## 1. สร้าง Workbook ใหม่

```python
# 📗 สร้าง Workbook ใหม่ — openpyxl.Workbook() สร้าง workbook เปล่าในหน่วยความจำ
# .active = sheet เริ่มต้นที่ถูกสร้างอัตโนมัติ (ปกติชื่อ "Sheet")
# .sheetnames = รายชื่อ sheet ทั้งหมดใน workbook
workbook = openpyxl.Workbook()
print("hello world")
print(f"สร้าง Workbook แล้ว")
print(f"Sheet ที่ใช้งาน: {workbook.active.title}")
print(f"Sheet ทั้งหมด: {workbook.sheetnames}")
```

## 2. สร้างและตั้งชื่อ Worksheet

```python
# ✏️ เปลี่ยนชื่อ sheet เริ่มต้นด้วย .title และสร้าง sheet ใหม่
# .title = property สำหรับอ่าน/เปลี่ยนชื่อ sheet
# .create_sheet("ชื่อ") = เพิ่ม sheet ใหม่ต่อท้าย
ws = workbook.active
ws.title = "Sales Data"
print(f"เปลี่ยนชื่อ sheet เป็น: {ws.title}")

ws2 = workbook.create_sheet("Summary")
ws3 = workbook.create_sheet("Charts")
print(f"Sheet ทั้งหมด: {workbook.sheetnames}")
```

## 3. เปลี่ยนสีแท็บ Worksheet

```python
# 🎨 เปลี่ยนสีแท็บของ Worksheet ด้วยรหัสสี hex (6 ตัวอักษร ไม่มี #)
# สีแท็บคือแถบสีด้านล่างของ Excel
# FF0000 = แดง, 00FF00 = เขียว, 0000FF = น้ำเงิน
ws.sheet_properties.tabColor = "FF0000"   # แดง
ws2.sheet_properties.tabColor = "00FF00"  # เขียว
ws3.sheet_properties.tabColor = "0000FF"  # น้ำเงิน
print("ตั้งสีแท็บแล้ว: Sales Data=แดง, Summary=เขียว, Charts=น้ำเงิน")
```

## 4. เพิ่มข้อมูลใน Worksheet

```python
# ✍️ เพิ่มข้อมูล — 3 วิธีเขียนข้อมูลลงเซลล์:
# วิธีที่ 1: ws['A1'] = ค่า  (อ้างอิงเซลล์แบบ Excel)
# วิธีที่ 2: ws.cell(แถว, คอลัมน์, ค่า)  (ใช้เลขแถว/คอลัมน์ เริ่มจาก 1)
ws['A1'] = 'Name'
ws['B1'] = 'Sales'
ws['C1'] = 'Region'
ws['D1'] = 'Date'

ws.cell(row=2, column=1, value='Alice')
ws.cell(row=2, column=2, value=5000)
ws.cell(row=2, column=3, value='East')
ws.cell(row=2, column=4, value='2024-01-15')
print("เพิ่มข้อมูลลงเซลล์แล้ว")
```

```python
# ➕ วิธีที่ 3: ws.append(แถว) — เพิ่มแถวทั้งแถวที่ตำแหน่งว่างถัดไป
# นี่คือวิธีที่ง่ายที่สุดในการเพิ่มข้อมูลหลายแถวตามลำดับ
# .max_row บอกจำนวนแถวทั้งหมดที่มีใน sheet
data = [
    ['Bob', 7500, 'West', '2024-02-20'],
    ['Charlie', 3200, 'North', '2024-03-10'],
    ['Diana', 8900, 'South', '2024-04-05'],
    ['Eve', 6100, 'East', '2024-05-18'],
    ['Frank', 4300, 'West', '2024-06-22'],
]

for row in data:
    ws.append(row)

print(f"เพิ่ม {len(data)} แถว รวมทั้งหมด: {ws.max_row} แถว")
```

## 5. อ่านข้อมูลจาก Worksheet

```python
# 📖 อ่านข้อมูล — สองวิธีในการอ่านเซลล์:
# ws['A1'].value — อ้างอิงแบบ Excel
# ws.cell(แถว, คอลัมน์).value — ใช้เลขแถว/คอลัมน์
print(f"เซลล์ A2: {ws['A2'].value}")
print(f"เซลล์ C2: {ws.cell(row=2, column=3).value}")
```

```python
# 📋 อ่านข้อมูลทั้งหมดใน worksheet
# .max_row / .max_column = ขนาดของพื้นที่ที่ใช้งาน
# ws.values คือ generator ที่ส่งแต่ละแถวกลับมาเป็น tuple
print(f"แถว: {ws.max_row}, คอลัมน์: {ws.max_column}")
print()

for row in ws.values:
    print(row)
```

## 6. การใช้ iter_rows() และ iter_cols()

```python
# 🔄 iter_rows() — วนลูปข้อมูลทีละแถว
# min_row/max_row = แถวที่ต้องการ (เริ่มจาก 1)
# min_col/max_col = คอลัมน์ที่ต้องการ
# values_only=True = ส่งกลับแค่ค่า แทนที่จะเป็น Cell objects
print("=== วนลูปทีละแถว ===")
for row in ws.iter_rows(min_row=1, max_row=4, min_col=1, max_col=3, values_only=True):
    print(row)
```

```python
# 🔄 iter_cols() — วนลูปข้อมูลทีละคอลัมน์
# พารามิเตอร์เหมือน iter_rows แต่ส่งกลับแต่ละคอลัมน์เป็น tuple
# มีประโยชน์เมื่อต้องประมวลผลทีละคอลัมน์
print("=== วนลูปทีละคอลัมน์ ===")
for col in ws.iter_cols(min_row=1, max_row=4, min_col=1, max_col=3, values_only=True):
    print(col)
```

```python
# 🎯 อ่านเฉพาะบางคอลัมน์ — ใช้ max_col เพื่อจำกัดคอลัมน์
# min_row=2 ข้ามแถวหัวตาราง
# แตกค่าแต่ละแถวเป็นตัวแปร (name, sales) เพื่อให้โค้ดอ่านง่าย
print("=== ชื่อและยอดขาย ===")
for row in ws.iter_rows(min_row=2, max_col=2, values_only=True):
    name, sales = row
    print(f"  {name}: ${sales:,}")
```

## 7. เข้าถึงแถวและคอลัมน์ด้วย Index

```python
# 📐 เข้าถึงช่วงเซลล์ด้วย slice เช่น ws['A1':'C3']
# ส่งกลับ tuple ซ้อน tuple — ชั้นนอก = แถว, ชั้นใน = เซลล์
# .coordinate = ที่อยู่เซลล์ (เช่น A1), .value = เนื้อหาเซลล์
cell_range = ws['A1':'C3']
for row in cell_range:
    for cell in row:
        print(f"{cell.coordinate}: {cell.value}", end="  ")
    print()
```

```python
# 📊 เข้าถึงคอลัมน์ทั้งหมด — ws['B'] ส่งกลับเซลล์ทั้งหมดในคอลัมน์ B
# รวมทุกแถวตั้งแต่บนสุดถึงล่างสุดของพื้นที่ที่ใช้งาน
col_b = ws['B']
print("ค่าในคอลัมน์ B:")
for cell in col_b:
    print(f"  {cell.value}")
```

## 8. บันทึก Workbook

```python
# 💾 บันทึก workbook ลงดิสก์
# .save() เขียน workbook ที่อยู่ในหน่วยความจำเป็นไฟล์ .xlsx จริง
# ⚠️ ไม่มีอะไรถูกบันทึกจนกว่าจะเรียก .save()!
workbook.save('openpyxl_essential_tips.xlsx')
print("✅ บันทึก Workbook เป็น 'openpyxl_essential_tips.xlsx' แล้ว")
```

## 9. เปิด Workbook ที่มีอยู่แล้ว

```python
# 📂 เปิด Workbook ที่มีอยู่แล้ว — load_workbook() อ่านไฟล์ .xlsx เข้าหน่วยความจำ
# สามารถเข้าถึง sheet ด้วยชื่อ: wb['ชื่อ Sheet']
# ใช้วิธีนี้เมื่อต้องการแก้ไขหรืออ่านไฟล์ที่มีอยู่แล้ว
wb_loaded = openpyxl.load_workbook('openpyxl_essential_tips.xlsx')
print(f"Sheet ใน workbook ที่เปิด: {wb_loaded.sheetnames}")

ws_loaded = wb_loaded['Sales Data']
for row in ws_loaded.iter_rows(min_row=1, max_row=3, values_only=True):
    print(row)
```

### โจทย์ที่ 1: สร้าง Workbook ข้อมูลนักเรียน 🟢
1. สร้าง Workbook ใหม่
2. ตั้งชื่อ sheet เป็น `"รายชื่อ"`
3. ใส่ header: `รหัส`, `ชื่อ`, `วิชา`, `คะแนน`
4. เพิ่มข้อมูลนักเรียน 5 คน โดยใช้ `ws.append()`
5. บันทึกเป็น `student_records.xlsx`

> 💡 Hint: ใช้ `Workbook()`, `ws.title`, `ws.append()`, `wb.save()`

```python
# โจทย์ที่ 1: สร้าง Workbook ข้อมูลนักเรียน
# เขียนโค้ดของคุณที่นี่ 👇
from openpyxl import Workbook

wb = Workbook()
ws = wb.active
ws.title= "รายชื่อ"

ws.append(["รหัส", "ชื่อ", "วิชา", "คะแนน"])

students = [
    ['001', 'สมชาย', 'คณิตศาสตร์', 85],
    ['002', 'สมหญิง', 'วิทยาศาสตร์', 92],
    ['003', 'สมปอง', 'ภาษาอังกฤษ', 78],
    ['004', 'สมศรี', 'คณิตศาสตร์', 88],
    ['005', 'สมบูรณ์', 'วิทยาศาสตร์', 95],
]

for student in students:
    ws.append(student)

wb.save('student_records.xlsx')
print("✅ สร้างไฟล์ 'student_records.xlsx' แล้ว")
```

### โจทย์ที่ 2: อ่านข้อมูลและค้นหา 🟡
1. เปิดไฟล์ `student_records.xlsx` ด้วย `load_workbook()`
2. อ่านข้อมูลทั้งหมดด้วย `iter_rows(values_only=True)`
3. หานักเรียนที่คะแนน **มากกว่า 80** แล้ว print ชื่อและคะแนน
4. คำนวณ **คะแนนเฉลี่ย** ของทุกคน

> 💡 Hint: ใช้ `load_workbook()`, `iter_rows(min_row=2, values_only=True)`

```python
# โจทย์ที่ 2: อ่านข้อมูลและค้นหา
# เขียนโค้ดของคุณที่นี่ 👇
from openpyxl import load_workbook

wb = load_workbook('student_records.xlsx')
ws = wb['รายชื่อ']

# อ่านข้อมูลทั้งหมด
scores = []
print('📋 นักเรียนที่คะแนนมากกว่า 80:')
for row in ws.iter_rows(min_row=2, values_only=True):
    rid, name, subject, score = row
    scores.append(score)
    if score > 80:
        print(f'  ✅ {name} — {score} คะแนน ({subject})')

avg = sum(scores) / len(scores)
print(f'\n💯 คะแนนเฉลี่ย: {avg:.1f}')
```

### โจทย์ที่ 3: เพิ่ม Sheet ใหม่พร้อมสรุป 🔴
1. เปิดไฟล์ `student_records.xlsx`
2. เพิ่ม sheet ใหม่ชื่อ `"สรุป"` โดยตั้ง tabColor เป็นสีเขียว (`"00FF00"`)
3. ใน sheet "สรุป" ใส่:
   - A1: `"จำนวนนักเรียน"`, B1: จำนวนนักเรียนทั้งหมด
   - A2: `"คะแนนเฉลี่ย"`, B2: คะแนนเฉลี่ย
   - A3: `"คะแนนสูงสุด"`, B3: คะแนนสูงสุด
4. บันทึกไฟล์

> 💡 Hint: อ่านข้อมูลจาก sheet "รายชื่อ" ก่อน แล้วคำนวณใส่ sheet "สรุป"

```python
# โจทย์ที่ 3: เพิ่ม Sheet ใหม่พร้อมสรุป
# เขียนโค้ดของคุณที่นี่ 👇
from openpyxl import load_workbook

wb = load_workbook('student_records.xlsx')
ws_data = wb['รายชื่อ']

# อ่านคะแนนทั้งหมด
scores = [row[3] for row in ws_data.iter_rows(min_row=2, values_only=True)]

# สร้าง sheet สรุป
ws_summary = wb.create_sheet('สรุป')
ws_summary.sheet_properties.tabColor = '00FF00'

ws_summary['A1'] = 'จำนวนนักเรียน'
ws_summary['B1'] = len(scores)
ws_summary['A2'] = 'คะแนนเฉลี่ย'
ws_summary['B2'] = round(sum(scores) / len(scores), 1)
ws_summary['A3'] = 'คะแนนสูงสุด'
ws_summary['B3'] = max(scores)

wb.save('student_records.xlsx')
print('✅ เพิ่ม sheet สรุป สำเร็จ!')
print(f'  จำนวน: {len(scores)} | เฉลี่ย: {sum(scores)/len(scores):.1f} | สูงสุด: {max(scores)}')
```
